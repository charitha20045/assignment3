```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'Sales': [150, 200, 250, 300, 350, 400],
    'Expenses': [100, 150, 200, 250, 300, 350]
}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 5))

plt.bar(df['Month'], df['Sales'], label='Sales', color='blue', alpha=0.7)
plt.bar(df['Month'], df['Expenses'], label='Expenses', color='orange', alpha=0.7)

plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Monthly Sales and Expenses')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Month'], df['Sales'], label='Sales', marker='o', color='blue')
plt.plot(df['Month'], df['Expenses'], label='Expenses', marker='o', color='orange')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Monthly Sales and Expenses')
plt.legend()
plt.show()
```
